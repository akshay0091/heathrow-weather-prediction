```python
df.drop(['SNWD (Inches)','SNOW (Inches)'] ,axis=1,inplace=True)
```

```python
df['PRCP (Inches)'].fillna(method="ffill",inplace=True)
```

```python
df['TMIN (Degrees Fahrenheit)'].fillna(method="ffill",inplace=True)
```

```python
df['TMAX (Degrees Fahrenheit)'].fillna(method="ffill",inplace=True)
```

```python
df.dtypes
```

```python
df.set_index('Date',inplace=True)
```

```python
df.index
```

```python
df.index= pd.to_datetime(df.index)
df.dtypes
```

```python
df.index.year
```

```python
df.rename(columns={'TAVG (Degrees Fahrenheit)':'TAVG'},inplace=True)
df.rename(columns={'TMAX (Degrees Fahrenheit)':'TMAX'},inplace=True)
df.rename(columns={'TMIN (Degrees Fahrenheit)':'TMIN'},inplace=True)
df.rename(columns={'PRCP (Inches)':'PRCP'},inplace=True)
```

```python
df.head()
```

```python
df[['TMAX','TMIN']].plot(figsize=(15,10))
```

```python
df[['PRCP']].plot(figsize=(15,10))
```

```python
df['target']=df.shift(-1)['TMAX']
```

```python
df
```

```python
df=df.iloc[:-1,:].copy()
df
```

```python
train=df.loc[:'2020-01-01']
test=df.loc['2020-01-01':]
```

```python
train
```

```python
test
```

```python
from sklearn.linear_model import Ridge

reg = Ridge(alpha=.1)
predictors = ["PRCP", "TMAX", "TMIN"]
```

```python
train[predictors]
```

```python
reg.fit(train[predictors], train["target"])
```

```python
predictions = reg.predict(test[predictors])
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(test["target"], predictions)
```

```python
combined = pd.concat([test["target"], pd.Series(predictions, index=test.index)], axis=1)
combined.columns = ["actual", "predictions"]
combined
```

```python
combined.plot(figsize=(15,10))
```
